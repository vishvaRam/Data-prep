```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python

# Load the dataset
dataset = load_dataset("Vishva007/RBI-Circular-QA-Dataset")

# Convert to Pandas DataFrames for easier EDA
train_df = pd.DataFrame(dataset['train'])
eval_df = pd.DataFrame(dataset['eval'])
```

```python

print("--- EDA for TRAIN Split ---")

# 1. Distribution of 'regulation_area'
plt.figure(figsize=(12, 6))
sns.countplot(data=train_df, y='regulation_area', order=train_df['regulation_area'].value_counts().index, palette='viridis')
plt.title('Distribution of Regulation Areas (Train Split)')
plt.xlabel('Count')
plt.ylabel('Regulation Area')
plt.show()

# Display value counts for clarity
print("\nRegulation Area Value Counts (Train Split):")
print(train_df['regulation_area'].value_counts())
```

```python

# 2. Distribution of 'applicable_to'
plt.figure(figsize=(12, 6))
sns.countplot(data=train_df, y='applicable_to', order=train_df['applicable_to'].value_counts().index, palette='plasma')
plt.title('Distribution of Applicable To Entities (Train Split)')
plt.xlabel('Count')
plt.ylabel('Applicable To')
plt.show()

print("\nApplicable To Value Counts (Train Split):")
print(train_df['applicable_to'].value_counts())
```

```python


# 3. Distribution of 'category'
plt.figure(figsize=(12, 6))
sns.countplot(data=train_df, y='category', order=train_df['category'].value_counts().index, palette='cividis')
plt.title('Distribution of Question Categories (Train Split)')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()

print("\nQuestion Category Value Counts (Train Split):")
print(train_df['category'].value_counts())
```

```python

# 4. Distribution of 'estimated_difficulty'
plt.figure(figsize=(8, 5))
sns.countplot(data=train_df, x='estimated_difficulty', order=train_df['estimated_difficulty'].value_counts().index, palette='magma')
plt.title('Distribution of Estimated Difficulty (Train Split)')
plt.xlabel('Estimated Difficulty')
plt.ylabel('Count')
plt.show()

print("\nEstimated Difficulty Value Counts (Train Split):")
print(train_df['estimated_difficulty'].value_counts())
```

```python

# 5. Length distribution of 'chunks_text', 'question', 'answer'
train_df['chunk_length'] = train_df['chunks_text'].apply(len)
train_df['question_length'] = train_df['question'].apply(len)
train_df['answer_length'] = train_df['answer'].apply(len)

plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
sns.histplot(train_df['chunk_length'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Chunk Text Length (Train Split)')
plt.xlabel('Length of Chunk Text')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
sns.histplot(train_df['question_length'], bins=50, kde=True, color='lightcoral')
plt.title('Distribution of Question Length (Train Split)')
plt.xlabel('Length of Question')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
sns.histplot(train_df['answer_length'], bins=50, kde=True, color='lightgreen')
plt.title('Distribution of Answer Length (Train Split)')
plt.xlabel('Length of Answer')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

print("\nDescriptive Statistics for Lengths (Train Split):")
print(train_df[['chunk_length', 'question_length', 'answer_length']].describe())
```

```python

# 6. Check 'is_table' distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=train_df, x='is_table', palette='pastel')
plt.title('Distribution of is_table (Train Split)')
plt.xlabel('Is Table Data')
plt.ylabel('Count')
plt.show()

print("\nIs Table Value Counts (Train Split):")
print(train_df['is_table'].value_counts())
```

```python

# 7. Most common key topics
# Flatten the list of key_topics and count occurrences
all_key_topics = [topic.strip() for sublist in train_df['key_topics'].str.split(',') for topic in sublist if topic.strip()]
key_topic_counts = Counter(all_key_topics)
```

```python

# Get the most common 20 topics
most_common_topics = key_topic_counts.most_common(20)
topics_df = pd.DataFrame(most_common_topics, columns=['Topic', 'Count'])

plt.figure(figsize=(12, 7))
sns.barplot(data=topics_df, y='Topic', x='Count', palette='crest')
plt.title('Top 20 Most Common Key Topics (Train Split)')
plt.xlabel('Count')
plt.ylabel('Key Topic')
plt.show()

print("\nTop 20 Most Common Key Topics (Train Split):")
for topic, count in most_common_topics:
    print(f"- {topic}: {count}")
```

```python

print("\n--- EDA for EVAL Split ---")
# For the eval split, we'll primarily look at the distribution of original features and also the impact of rephrasing

# 1. Distribution of 'regulation_area' (Eval Split) - Should be similar to train if random sampling
plt.figure(figsize=(12, 6))
sns.countplot(data=eval_df, y='regulation_area', order=eval_df['regulation_area'].value_counts().index, palette='viridis')
plt.title('Distribution of Regulation Areas (Eval Split)')
plt.xlabel('Count')
plt.ylabel('Regulation Area')
plt.show()

print("\nRegulation Area Value Counts (Eval Split):")
print(eval_df['regulation_area'].value_counts())
```

```python

# 2. Length distribution of original vs rephrased questions/answers in eval split
eval_df['question_length_original'] = eval_df['question'].apply(len)
eval_df['question_length_rephrased'] = eval_df['rephrased_question'].apply(len)
eval_df['answer_length_original'] = eval_df['answer'].apply(len)
eval_df['answer_length_rephrased'] = eval_df['rephrased_answer'].apply(len)

plt.figure(figsize=(18, 8))

plt.subplot(2, 2, 1)
sns.histplot(eval_df['question_length_original'], bins=20, kde=True, color='purple', label='Original')
sns.histplot(eval_df['question_length_rephrased'], bins=20, kde=True, color='orange', label='Rephrased')
plt.title('Distribution of Question Lengths (Eval Split)')
plt.xlabel('Length of Question')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(2, 2, 2)
sns.histplot(eval_df['answer_length_original'], bins=20, kde=True, color='purple', label='Original')
sns.histplot(eval_df['answer_length_rephrased'], bins=20, kde=True, color='orange', label='Rephrased')
plt.title('Distribution of Answer Lengths (Eval Split)')
plt.xlabel('Length of Answer')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(2, 2, 3)
sns.boxplot(data=eval_df[['question_length_original', 'question_length_rephrased']])
plt.title('Box Plot of Question Lengths (Eval Split)')
plt.ylabel('Length')
plt.xticks(ticks=[0, 1], labels=['Original', 'Rephrased'])

plt.subplot(2, 2, 4)
sns.boxplot(data=eval_df[['answer_length_original', 'answer_length_rephrased']])
plt.title('Box Plot of Answer Lengths (Eval Split)')
plt.ylabel('Length')
plt.xticks(ticks=[0, 1], labels=['Original', 'Rephrased'])

plt.tight_layout()
plt.show()

print("\nDescriptive Statistics for Lengths (Eval Split - Original vs Rephrased):")
print("Original Question Length:")
print(eval_df['question_length_original'].describe())
print("\nRephrased Question Length:")
print(eval_df['question_length_rephrased'].describe())
print("\nOriginal Answer Length:")
print(eval_df['answer_length_original'].describe())
print("\nRephrased Answer Length:")
print(eval_df['answer_length_rephrased'].describe())
```
